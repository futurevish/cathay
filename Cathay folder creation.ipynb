```python
import pandas as pd
```

```python
import os
import re
import shutil

def sort_files_by_gender(source_dir, target_base_dir, id_list,gender):
    # Create gender category folders
    print(target_base_dir,gender)
    
    
    # Compile regex pattern for matching IDs
    pattern = '|'.join(map(re.escape, id_list))
    
    # Iterate through files in source directory
    for filename in os.listdir(source_dir):
        print(filename)
        match = re.search(pattern, filename)
        if match:
            matched_id = match.group()
            # Determine gender based on ID prefix
            
            # Source and destination paths
            src_path = os.path.join(source_dir, filename)
            
            directory_name=target_base_dir+'_'+str(gender)
            if not os.path.exists(directory_name):
                os.mkdir(directory_name)
            if not os.path.exists(os.path.join(directory_name, 'Sensor Data')):
                os.mkdir(os.path.join(directory_name, 'Sensor Data'))
            dst_path = os.path.join(directory_name, 'Sensor Data')
            
            # Move file to appropriate folder
            shutil.copy(src_path, dst_path)
            if not os.path.exists(os.path.join(directory_name, 'survey_data')):
                os.mkdir(os.path.join(directory_name, 'survey_data'))
            shutil.copy("./CX_LAX_EN/survey_data/MERGED_SURVEY_RESPONSE_MATRIX.txt",  os.path.join(directory_name, 'survey_data'))
            print(f"Moved {filename} to {gender} folder")

# Example usage
study = "CX_LAX_EN"
source_directory = f"./{study}/Sensor Data/"

city = 'LAX'
segment = 'Gender'
target_directory = f"{study}_{segment}"

resp_track_df=pd.read_excel("C:/Users/FutureProofInsights/OneDrive - Future Proof Media/Documents/Cathay Pacific/Respondent Tracking/Respondent Tracking_AllMarkets2.xlsx",sheet_name='Full Panel Overview')
for gender in resp_track_df[(resp_track_df['CITY']==city)][segment].unique():
    id_list=resp_track_df[(resp_track_df[segment]==gender) & (resp_track_df['CITY']==city)]['Imotions ID '].unique()# Add your ID list her
    sort_files_by_gender(source_directory, target_directory, id_list,gender)
```
